```python
year = dataset['iyear'].unique()
years_count = dataset['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation=90)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=25)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

```python

plt.subplots(figsize=(20,10))
sns.barplot(dataset['country_txt'].value_counts()[:10].index,dataset['country_txt'].value_counts()[:10].values,palette='tab10')
plt.title('Top Countries Affected')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks()
plt.show()
```

```python

dataset.plot(kind = 'scatter', x = 'nkill', y = 'nwound', alpha = 0.5, color = 'red', figsize = (20,20), fontsize=15)
plt.xlabel('Kill', fontsize=15)
plt.ylabel('Wound', fontsize=15)
plt.title('Kill - Wound Scatter Plot',fontsize=50)
plt.show()
```

```python
##Frequency of Terrorist Actions in Customized Region
#Let's analyze in Middle East and North Africa

middleEastData = dataset[dataset['region'] == 10]
middleEastData.iyear.plot(kind = 'hist', bins = 30, figsize = (20,20), color = 'blue', fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.title('Frequency of Middle East & North Africa Terrorism Actions by Years')
plt.show()
```

```python
killData = dataset.loc[:,'nkill']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))# drop the NaN values
```

```python

attackData = dataset.loc[:,'attacktype1':'attacktype1_txt']
# attackData
typeKillData = pd.concat([attackData, killData], axis=1)
```

```python
typeKillFormatData = typeKillData.pivot_table(columns='attacktype1_txt', values='nkill', aggfunc='sum')
typeKillFormatData
```

```python

plt.subplots(figsize=(20,10))
sns.barplot(dataset['attacktype1_txt'].value_counts()[:10].index,dataset['attacktype1_txt'].value_counts()[:10].values,palette='tab10')
plt.title('Top Countries Affected')
plt.xlabel('Most attack types')
plt.ylabel('Count')
plt.xticks()
plt.show()
```

```python

countryData = dataset.loc[:,'country':'country_txt']
# countyData
countryKillData = pd.concat([countryData, killData], axis=1)
```

```python
countryKillFormatData = countryKillData.pivot_table(columns='country_txt', values='nkill', aggfunc='sum')
countryKillFormatData
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size

labels = countryKillFormatData.columns.tolist()
labels = labels[:50] #50 bar provides nice view
index = np.arange(len(labels))
transpoze = countryKillFormatData.T
values = transpoze.values.tolist()
values = values[:50]
values = [int(i[0]) for i in values] # convert float to int
colors = ['red', 'green', 'blue', 'purple', 'yellow', 'brown', 'black', 'gray', 'magenta', 'orange'] # color list for bar chart bar color 
fig, ax = plt.subplots(1, 1)
ax.yaxis.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
plt.bar(index, values, color = colors, width = 0.9)
plt.ylabel('Killed People', fontsize=15)
plt.xticks(index, labels, fontsize=12, rotation=90)
plt.title('Number of people killed by countries')
# print(fig_size)
plt.show()
```
